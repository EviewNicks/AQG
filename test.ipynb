```python
# Load environment variables
load_dotenv()

# print("openAi version : ", OpenAI._version_)

llm_openrouter = ChatOpenAI(
    model=os.getenv("ZAI_MODEL"),
    api_key=os.getenv("ZAI_API_KEY"),
    base_url="https://api.z.ai/api/paas/v4",
    # seed= 365,
    # temperature= 7 ,
    # max_completion_tokens= 100
)

# Test simple completion
print("🧪 Testing GLMs connection...")

try:
    message = HumanMessage(content="Please write a short fairy tale. Keep it short.")

    response = llm_openrouter.invoke([message])

    print("📖 Fairy Tale:")
    print(response.content)
    print(f"\n📊 Tokens: {response.response_metadata.get('token_usage', 'N/A')}")

    print(f"\n✅ Koneksi berhasil!")

except Exception as e:
    print(f"\n❌ Error: {e}")
    print("\nTroubleshooting:")
    print("1. Cek API key di file .env")
    print("2. Pastikan internet connection aktif")
    print("3. Verify API key valid di dashboard Z.AI")
```

```python
# ============================================================
# Test: langchain-openrouter integration
# Menggunakan OPENROUTER_API_KEY dan OPENROUTER_MODEL dari .env
# Install: pip install langchain-openrouter
# ============================================================
from dotenv import load_dotenv
import os
from langchain_openrouter import ChatOpenRouter
from langchain_core.messages import HumanMessage, SystemMessage

load_dotenv(override=True)

# Inisialisasi client
llm_or = ChatOpenRouter(
    model=os.getenv("OPENROUTER_MODEL", "z-ai/glm-4.5-air:free"),
    api_key=os.getenv("OPENROUTER_API_KEY"),
    temperature=float(os.getenv("OPENROUTER_TEMPERATURE", "0.7")),
    max_tokens=int(os.getenv("OPENROUTER_MAX_TOKENS", "2000")),
)

print(f"🧪 Testing OpenRouter — model: {os.getenv('OPENROUTER_MODEL')}")

try:
    messages = [
        SystemMessage(content="Kamu adalah asisten yang membantu."),
        HumanMessage(content="Tulis satu kalimat tentang Python programming."),
    ]
    response = llm_or.invoke(messages)

    print("💬 Response:")
    print(response.content)
    print(f"\n📊 Usage: {response.usage_metadata}")
    print(f"📋 Model: {response.response_metadata.get('model_name', 'N/A')}")
    print("\n✅ OpenRouter koneksi berhasil!")

except Exception as e:
    print(f"\n❌ Error: {e}")
    print("\nTroubleshooting:")
    print("1. Pastikan langchain-openrouter sudah terinstall: pip install langchain-openrouter")
    print("2. Cek OPENROUTER_API_KEY di file .env")
    print("3. Verify model name valid di https://openrouter.ai/models")
```
